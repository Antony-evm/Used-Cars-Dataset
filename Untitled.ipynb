```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import opendatasets as od
import plotly.figure_factory as ff
import plotly.graph_objects as go

pd.options.plotting.backend = "plotly"
```

```python
od.download('https://www.kaggle.com/datasets/shivanink8/used-cars-dataset')
```

```python
df = pd.read_csv('used-cars-dataset/used_cars.csv')
```

```python
display(df.head(5))
df.shape
```

```python
df.isnull().sum()
```

```python
for i in df.columns:
    display(df[i].value_counts())
```

```python
df = df.drop(['link','condition'],axis=1)
```

```python
df = df.drop(['num_seat','type','num_cylinder','doors'],axis=1)
```

```python
df.describe()
```

```python
variables = df.drop('price',axis=1).columns[3:]
variables
```

```python
df['speed_levels'] = df['speed_levels'].fillna(df['speed_levels'].mean())
```

```python
for i in [j for j in variables if df[j].dtype!='O']:
    fig = px.scatter(df, x=i, y='price', color='price', size=i)
    fig.update_layout(title=f'Scatter Plot of {i} vs. Price Colored by Price', xaxis_title=i, yaxis_title='price')
    fig.show()
```

```python
for i in [j for j in variables if df[j].dtype!='O']:
    fig = px.box(df, y=i)
    fig.update_layout(title=i,yaxis_title=i, template="plotly_white")
    fig.update_yaxes(showgrid=False)
    fig.update_xaxes(showgrid=False)
    fig.show()
```
